```python
meta = []
Ps = []
for (i,dm) in enumerate(tqdm(blazeit_dms, desc='dataset')):
    ds = BenchmarkDataset(dm)
    st = ds.make_split('60min')
    Pdfs = getPs(ds, st)
    for (k,P) in Pdfs.items():
        meta.append((ds.name, k))
        Ps.append(P.values)
```

```python
sps = []
vps = []
Ns = []
mus = []
for P in Ps:
    sps.append(P.sum(axis=0))
    vps.append(P.std(axis=0))
    mus.append(P.mean(axis=0))
    Ns.append((P > 0).sum(axis=0))
```

```python
dfs = []
for (m,sp,std,mu,n) in zip(meta,sps,vps,mus,Ns):
    spdf = pd.DataFrame({'value':sp}).assign(dataset=m[0], category=m[1], metric='sm_p').reset_index()
    stdf = pd.DataFrame({'value':std}).assign(dataset=m[0], category=m[1], metric='std_p').reset_index()
    mudf = pd.DataFrame({'value':mu}).assign(dataset=m[0], category=m[1], metric='mu_p').reset_index()
    ndf = pd.DataFrame({'value':n}).assign(dataset=m[0], category=m[1], metric='N').reset_index()

    dfs.append(mudf)
    dfs.append(stdf)
    dfs.append(ndf)
    dfs.append(spdf)
```

```python
mstdf = pd.concat(dfs)
```

```python
(ggplot(mstdf.reset_index())
  + geom_col(aes(x='index',y='value'))
  + facet_wrap(facets=['dataset', 'category','metric'],ncol=4,scales='free')
  + theme(figure_size=(4*3, len(meta)*3), subplots_adjust={'wspace':.2, 'hspace':.2})
)
```
